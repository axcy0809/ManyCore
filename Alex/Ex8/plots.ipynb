```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("dot_times.csv",sep = ";")
df
```

```python
plt.figure(figsize=(13,10))
plt.loglog(df["N"],df["CPU_time"], label = "CPU_Time", marker='x')
plt.loglog(df["N"],df["VexCL_Time"], label = "VexCL_Time", marker='x')
plt.loglog(df["N"],df["ViennaCL_Time"], label = "ViennaCL_Time", marker='x')
plt.loglog(df["N"],df["BoostCompute_Time"], label = "BoostCompute_Time", marker='x')
plt.legend()
plt.grid(True, which="both",linewidth=0.7)
plt.title("Runtime dotProduct")
plt.xlabel("length of vector")
plt.ylabel("time[s]")
plt.savefig("Bilder/Runtime_dotProduct")
```
