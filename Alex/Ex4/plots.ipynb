```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = [1e3,1e4,1e5,2e5,4e5,6e5,8e5,1e6]
erg8 = [0.3456,0.5029, 2.1402,3.9415,7.9655,12.7002,17.7754, 23.8381]
erg16 = [0.7257, 1.068, 4.4817, 7.4766, 14.1127,22.5356,27.6329, 35.4391]

plt.plot(N,erg8, label = "K=8", marker='x')
plt.plot(N,erg16, label = "K=16", marker='x')
plt.legend()
plt.grid()
plt.title("Runtime vs different K")
plt.xlabel("length of vector")
plt.ylabel("time[s]")
plt.savefig("Bilder/runtime_vs_K")
```
