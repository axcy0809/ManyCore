```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
```

```python
df_shared = pd.read_csv("shared_times.csv", sep=";")
df_suffeld = pd.read_csv("shuffeld.csv", sep=";")
df_shared
```

```python
#mpl.style.use('seaborn')
plt.figure(figsize=(13,10))
plt.title("shared vs shuffeled")
plt.ylabel("time[s]")
plt.xlabel("size of vector")
plt.grid(True, which="both",linewidth=0.6)
plt.loglog(df_shared["N"],df_shared["Time_mmsszz_shared"], label = "sssmmmz_shared", marker="o")
plt.loglog(df_shared["N"],df_shared["Time_mmsszz_worp_shuffel"], label = "sssmmmz_shuffel", marker="o")
plt.legend()
plt.savefig("Bilder/shared_vs_shuffeled")
```

```python
#mpl.style.use('seaborn')
plt.figure(figsize=(13,10))
plt.title("shuffeled vs shuffeled_adapt")
plt.ylabel("time[s]")
plt.xlabel("size of vector")
plt.grid(True, which="both",linewidth=0.5)
plt.loglog(df_shared["N"],df_shared["Time_mmsszz_worp_shuffel"], label = "sssmmmz_shuffel", marker="o")
plt.loglog(df_suffeld["N"],df_suffeld["Time_mmsszz_worp_shuffel"], label = "sssmmmz_shuffel_adapt", marker="o")
plt.legend()
plt.savefig("Bilder/shuffeled_vs_shuffeled_adapt")
```

```python
#mpl.style.use('seaborn')
plt.figure(figsize=(13,10))
plt.title("shared_Dot vs CUBLAS_Dot")
plt.ylabel("time[s]")
plt.xlabel("size of vector")
plt.grid(True, which="both",linewidth=0.5)
plt.loglog(df_shared["N"],df_shared["Time_DotProd_shared"], label = "DotProd_shared", marker="o")
plt.loglog(df_shared["N"],df_shared["Time_CUBLAS_DotProduct"], label = "CUBLAS_DotProduct",marker="o")
plt.legend()
plt.savefig("Bilder/shared_Dot_vs_CUBLAS_Dot")
```
