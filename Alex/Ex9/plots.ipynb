```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_CPU = pd.read_csv("dot_CPU_times.csv",sep = ";")
df_Boost = pd.read_csv("dot_Boost_times.csv",sep = ";")
df_Vexcl = pd.read_csv("dot_VexCL_times.csv",sep = ";")
df_Viennacl = pd.read_csv("dot_ViennaCl_times.csv",sep = ";")
df_Thrust = pd.read_csv("dot_Thrust_times.csv",sep = ";")
df_myCUDA = pd.read_csv("dot_myCUDA_times.csv",sep = ";")
df_myopenCL = pd.read_csv("dot_myopenCL_times.csv",sep = ";")
df_myCUDA
```

```python
plt.figure(figsize=(13,10))
plt.loglog(df_CPU["N"],df_CPU["CPU_time"], label = "CPU_Time", marker='x')
plt.loglog(df_CPU["N"],df_Vexcl["VexCL_time"], label = "VexCL_Time", marker='x')
plt.loglog(df_CPU["N"],df_Viennacl["ViennaCl_time"], label = "ViennaCL_Time", marker='x')
plt.loglog(df_CPU["N"],df_Boost["Boost_time"], label = "BoostCompute_Time", marker='x')
plt.loglog(df_CPU["N"],df_Thrust["Thrust_time"], label = "Thrust_time", marker='x')
plt.loglog(df_CPU["N"],df_myopenCL["myopenCL_time"], label = "myopenCL_time", marker='x')
plt.loglog(df_CPU["N"],df_myCUDA["myCUDA_time"], label = "myCUDA_time", marker='x')
plt.legend()
plt.grid(True, which="both",linewidth=0.7)
plt.title("Runtime dotProduct")
plt.xlabel("length of vector")
plt.ylabel("time[s]")
plt.savefig("Bilder/Runtime_dotProduct")
```
