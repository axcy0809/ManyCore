```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = [10,100,1000,10000,100000,1000000,10000000,100000000]
timeAll = [7.80125,7.58165,7.8174,8.22262,8.14841,7.9138,12.0587,29.637]
plt.plot(N,timeAll)
plt.grid()
plt.ylabel('time[ms]')
plt.xlabel('length of array')
plt.title("Time to allocate and free")
plt.savefig("Bilder/Time_to_allocate_and_free")
```

```python
N = [100,300,1000,3000,10000,30000,100000]
timeAll = [21.6434,21.2801,20.7999,20.12,20.1156,20.115,22.5097]
plt.plot(N,timeAll)
plt.grid()
plt.ylabel('time[ms]')
plt.xlabel('length of array')
plt.title("Time to Sum")
plt.savefig("Bilder/Time_to_Sum")
```

```python
N = [16,32,64,128,256,512,1024]
timeAll = [179.16, 98.476, 81.672, 63.587, 80.033, 96.152,74.541]
plt.loglog(N,timeAll)
plt.grid()
plt.ylabel('time[ms]')
plt.xlabel('length of array')
plt.title("Time Diff size")
plt.savefig("Bilder/Diff_size")
```
